```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
school_data_complete.count()
```

```python
organized_data = school_data_complete.groupby(["school_name"])
school_summary_table = pd.DataFrame({"School Name":school_data["school_name"], "School Type":school_data["type"], "Total Budget":school_data["budget"]})
school_summary_table = school_summary_table.set_index("School Name")
school_summary_table = school_summary_table.sort_values("School Name")

school_summary_table["Total Students"] = organized_data["Student ID"].count()
school_summary_table["Budget / Student"] = school_summary_table["Total Budget"]/school_summary_table["Total Students"]
school_summary_table["Av Math Score"] = round(organized_data["math_score"].mean(), 2)
school_summary_table["Av Reading Score"] = round(organized_data["reading_score"].mean(), 2)
school_summary_table["% Passing Math"] = round((organized_data.apply(lambda x: (x["math_score"] >= 70).sum()) / school_summary_table["Total Students"]) * 100, 2)
school_summary_table["% Passing Reading"] = round((organized_data.apply(lambda x: (x["reading_score"] >= 70).sum()) / school_summary_table["Total Students"]) * 100, 2)
school_summary_table["Overall Passing Rate"] = round((school_summary_table["% Passing Math"] + school_summary_table["% Passing Reading"]) / 2, 2)

print(school_summary_table)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
ninth = school_data_complete.loc[school_data_complete["grade"] == "9th"]
tenth = school_data_complete.loc[school_data_complete["grade"] == "10th"]
eleventh = school_data_complete.loc[school_data_complete["grade"] == "11th"]
twelfth = school_data_complete.loc[school_data_complete["grade"] == "12th"]

math_ninth = ninth.groupby("school_name")["math_score"].mean()
math_tenth = tenth.groupby("school_name")["math_score"].mean()
math_eleventh = eleventh.groupby("school_name")["math_score"].mean()
math_twelfth = twelfth.groupby("school_name")["math_score"].mean()

math_grades = pd.DataFrame({"Ninth":math_ninth, "Tenth":math_tenth, 
                             "Eleventh":math_eleventh, "Twelfth":math_twelfth})

math_grades = round(math_grades, 2)

math_grades
```
